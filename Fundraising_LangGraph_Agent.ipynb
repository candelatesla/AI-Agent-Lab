```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.1,
    )
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

emb = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create a Chroma vector store
db = Chroma(collection_name="fundraising_docs", embedding_function=emb)

# Add documents to the vector store
from tqdm import tqdm  # optional, for progress bar

batch_size = 5000  # safe batch size for Chroma
for i in tqdm(range(0, len(all_docs), batch_size)):
    batch = all_docs[i:i+batch_size]
    db.add_texts(batch)

```

```python
from typing import List
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage
from pydantic import BaseModel

class AgentState(BaseModel):
    messages: List[BaseMessage]
    retrieved_docs: List[str] = []

# Node to retrieve relevant documents
def retrieve_node(state: AgentState):
    query = state.messages[-1].content
    results = db.similarity_search(query, k=5)  # get top 5 docs for richer context
    retrieved = [d.page_content for d in results]
    return {"retrieved_docs": retrieved}

# Node to reason over retrieved docs
def reasoning_node(state: AgentState):
    context = "\n".join(state.retrieved_docs)
    user_msg = state.messages[-1].content

    prompt = f"""
You are an expert fundraising AI assistant. Use the context to answer the user's query.

Context:
{context}

Question: {user_msg}
"""
    response = llm.invoke(prompt)
    return {"messages": [AIMessage(content=response.content)]}
```

```python
from langgraph.graph import StateGraph, END

graph = StateGraph(AgentState)

graph.add_node("retrieve", retrieve_node)
graph.add_node("reason", reasoning_node)

graph.set_entry_point("retrieve")
graph.add_edge("retrieve", "reason")
graph.add_edge("reason", END)

final_graph = graph.compile()
```

```python
state = AgentState(messages=[HumanMessage(content="Which campaigns have the highest donor engagement?")])

result = final_graph.invoke(state)

print(result["messages"][-1].content)
```
